## movie analisis

first the movie data must be collected. movies were collected beforehand from unknown sources (the hard drives from 100% not pirated movies fell out of the sky). movies were missing the genres and usefull info. so the info was retrieved from ČSFD by a open source javascript project on github [here](https://github.com/bartholomej/node-csfd-api).

### scraper logic

The scraper first processes the movie names and srips all information like resolution, dabbing and all things that are not part of the tittle. Because it can miss some junk the scraper tries to than find the best tittle match on ČSFD and prompts the user whether its correct or not. The user can than change the movie id if the scraper guessed incorectly.
